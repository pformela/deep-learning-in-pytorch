```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
# manual softmax calculation in numpy

z = [1, 2, 3]

# compute the softmax result
num = np.exp(z)
den = np.sum(np.exp(z))
sigma = num / den

sigma, np.sum(sigma)
```

```python
# repeat with random integers
z = np.random.randint(-5, high=15, size=25)

print(z)

# compute the softmax result
num = np.exp(z)
den = np.sum(num)
sigma = num / den

# compare
plt.plot(z, sigma, 'ko')
plt.xlabel('Original number (z)')
plt.ylabel('Softmaxified number $\sigma$')
plt.title('$\sum\sigma$ = %g' %np.sum(sigma))
plt.show()
```

## Using pytorch

```python
# create an instance of the softmax activation class
softfun = nn.Softmax(dim=0)

# apply the data to that function
sigmaT = softfun(torch.Tensor(z))

print(sigmaT)
```

```python
# Compare results
plt.plot(sigma, sigmaT, 'ko')
plt.xlabel('"Manual" softmax')
plt.ylabel('Pytorch nn.Softmax')
plt.title(f'The two methods correlate at r={np.corrcoef(sigma, sigmaT)[0, 1]}')
plt.show()
```
