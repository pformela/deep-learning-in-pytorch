```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create a list of numbers to compute the mean and variance of
x = [1,2,4,6,5,4,0,-4,5,-2,6,10,-9,1,3,-6]
n = len(x)

# compute the population mean
popmean = np.mean(x)

# compute the sample mean
sample = np.random.choice(x, size=5, replace=True)
sampmean = np.mean(sample)

popmean, sampmean
```

```python
# n of experiments to run
n_experiments = 10000

sample_means = np.zeros(n_experiments)

for i in range(n_experiments):
  sample = np.random.choice(x, size=5, replace=True)
  sample_means[i] = np.mean(sample)

# show the results as a histogram
plt.hist(sample_means, bins=40, density=True)
plt.plot([popmean, popmean], [0, 0.3], 'm--')
plt.ylabel('Count')
plt.xlabel('Sample mean')
plt.show()
```
